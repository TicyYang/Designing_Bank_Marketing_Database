```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.sql import text

db_url = 'postgresql://username:password@localhost:5432/bank_marketing'
conn = create_engine(db_url).connect()

query = text('SELECT * FROM client LIMIT 5;')
result = conn.execute(query)

pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
query = text('SELECT * FROM campaign WHERE previous_outcome = true LIMIT 5;')
result = conn.execute(query)

pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
query = text('SELECT job, education, ROUND(AVG(contact_duration),2) AS avg_cont_dur \
              FROM client JOIN campaign ON client.id = campaign.client_id \
              GROUP BY job, education \
              ORDER BY avg_cont_dur DESC \
              LIMIT 5;')
result = conn.execute(query)

pd.DataFrame(result.fetchall(), columns=result.keys())
```

```python
conn.close()
```
